# Pandas - Filtering and Sorting Data

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
chipo.head()
```

```python
chipo.info()
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].str.replace('$','').astype(float)    #change to float
```

```python
chipo.info()
```

```python
lt10=chipo[chipo['item_price']> 10]
lt10.head()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.columns
```

```python
each_item=chipo[['item_name','item_price']].groupby('item_name')
print(list(each_item))
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name', ascending=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
max_price=chipo['item_price'].max()
print(chipo['quantity'][chipo['item_price']==max_price].sum())
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
print(chipo['quantity'][chipo['item_name']=='Veggie Salad Bowl'].sum())
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
print((chipo['quantity']>2)[chipo['item_name']=='Canned Soda'].sum())
```
